```python
import pandas as pd
```

```python
df_bz = pd.read_excel("E:\Code\shap_word\文件名称标准表.xlsx")
df_bz
```

```python
# 去除包含特定注释的行（非结构化数据行）
df_bz = df_bz[~df_bz['层名称'].str.contains('约束条件取值', na=False)]
df_bz.dropna(subset=['层名称','属性表名'],inplace=True)
# 按字段名称和字段代码去重（保留第一个出现的记录）
df_bz = df_bz.drop_duplicates(subset=['层名称', '属性表名'])

# 重置索引（可选）
df_bz = df_bz.reset_index(drop=True)

# 显示处理后的DataFrame
df_bz
```

```python
df_bz.to_excel("文件名称映射表.xlsx", index=False)
```
