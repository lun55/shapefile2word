```python
import pandas as pd
```

```python
df_bz = pd.read_excel("E:\Code\shap_word\字段标准表.xlsx")
df_bz
```

```python
# 去除包含特定注释的行（非结构化数据行）
df_bz = df_bz[~df_bz['字段名称'].str.contains('用于存储外挂的文档文件的路径|约束条件取值', na=False)]

# 按字段名称和字段代码去重（保留第一个出现的记录）
df_bz = df_bz.drop_duplicates(subset=['字段名称', '字段代码'])

# 重置索引（可选）
df_bz = df_bz.reset_index(drop=True)

# 显示处理后的DataFrame
df_bz
```

```python
df_bz.to_excel("标准映射编码表.xlsx",index=False)
```
